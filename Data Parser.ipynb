```python
import numpy as np
import cv2
import os
```

```python
def img_data_parse(filenames):
    img = cv2.imread(filenames, 0)
    #cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
curr_dir = os.getcwd()
join = os.path.join

img_dirs = ["Images_Background\\Images","Images_Drone\\Images"]
drones = []
background = []
```

```python
for i, folder in enumerate(img_dirs):
    path = join(curr_dir, folder)
    files = os.listdir(folder)
    for j, file in enumerate(files):
        print("Adding {} imgage {}".format(img_dirs[i],file))
        file = join(path, file)
        img = img_data_parse(file)
        if(i):
            drones.append(img)
        else:
            background.append(img)
```

```python
np.save('images_drone', np.array(drones))
np.save('images_background', np.array(background))
```

```python
background_resampled = []
drone_resampled = []
```

```python
from scipy import signal
from tqdm import tqdm
M = 2**9
num_samples = int(10**7/M)
print("Resample size:",num_samples)
print("Processing background RF..")
for sample in tqdm(background):
    resampled_L = signal.resample(sample[0], num_samples)
    resampled_H = signal.resample(sample[1], num_samples)
    background_resampled.append(np.array([resampled_L, resampled_H]))
print("Processing drone RF..")
for sample in tqdm(drone):
    resampled_L = signal.resample(sample[0], num_samples)
    resampled_H = signal.resample(sample[1], num_samples)
    drone_resampled.append(np.array([resampled_L, resampled_H]))
```

```python
from scipy import signal
from tqdm import tqdm
M = 2**11
num_samples = int(10**7/M)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

print("Processing background RF..")
for sample in tqdm(background):
    resampled_L = signal.resample(sample[0], num_samples)
    resampled_H = signal.resample(sample[1], num_samples)
    resampled_LH = np.array([resampled_L, resampled_H])
    resampled_LH = scaler.fit_transform(resampled_LH)
    background_resampled.append(resampled_LH)
print("Processing drone RF..")
for sample in tqdm(drone):
    resampled_L = signal.resample(sample[0], num_samples)
    resampled_H = signal.resample(sample[1], num_samples)
    resampled_LH = np.array([resampled_L, resampled_H])
    resampled_LH = scaler.fit_transform(resampled_LH)
    drone_resampled.append(resampled_LH)
```

```python
np.save("data/droneRF_resampled" ,np.array(drone_resampled))
np.save("data/backgroundRF_resampled" ,np.array(background_resampled))
```
