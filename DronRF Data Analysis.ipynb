```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
background = np.load("data/background_rf_LH.npy")
drone = np.load("data/drone_rf_LH.npy")
```

```python
num = random.randint(0, len(background)-1)
channel = 0
plt.plot(background[num][channel], label="background")
plt.plot(drone[num][channel],label="drone")
plt.legend(loc='upper right')
```

```python
def split_rf(x_data):
    low = []
    high = []
    for x in x_data:
        low.append(x[0].flatten().astype(np.float16))
        high.append(x[1].flatten().astype(np.float16))
    low = np.array(low)
    high = np.array(high)
    return [low, high]

drone = split_rf(drone)
background = split_rf(background)
```
