```python
import numpy as np
import cv2
import os
```

```python
def img_data_parse(filenames):
    img = cv2.imread(filenames, 0)
    #cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
curr_dir = os.getcwd()
join = os.path.join

img_dirs = ["Images_Background\\Images","Images_Drone\\Images"]
drones = []
background = []
```

```python
for i, folder in enumerate(img_dirs):
    path = join(curr_dir, folder)
    files = os.listdir(folder)
    for j, file in enumerate(files):
        print("Adding {} imgage {}".format(img_dirs[i],file))
        file = join(path, file)
        img = img_data_parse(file)
        if(i):
            drones.append(img)
        else:
            background.append(img)
```

```python
np.save('images_drone', np.array(drones))
np.save('images_background', np.array(background))
```
